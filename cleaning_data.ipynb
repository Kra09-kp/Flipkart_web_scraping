```python
def edit_value(index,processor,ram,os,storage,display,warranty):
    df.loc[index,'Processor'] = processor
    df.loc[index,'Ram'] = ram
    df.loc[index,'OS'] = os
    df.loc[index,'Storage'] = storage
    df.loc[index,'Display'] = display
    df.loc[index,'Warranty'] = warranty
```

```python
index = [118,165,329,494]
processor = ['Intel Core i7 Processor (12th Gen)','Intel Core i7 Processor (12th Gen)','Intel Core i5 Processor (12th Gen)','Intel Core i5 Processor (12th Gen)']
ram = ['16GB DDR4','16GB DDR4','8GB DDR4','16GB DDR4']
os = ['Windows 10','Windows 10','Windows 10','Windows 10']
storage = ['512 GB SSD','512 GB SSD','512 GB SSD','512 GB SSD']
display = [np.nan,np.nan,np.nan,np.nan]
warranty = [np.nan,np.nan,np.nan,np.nan]
```

```python
for i,p,r,o,s,d,w in zip(index,processor,ram,os,storage,display,warranty):
    edit_value(i,p,r,o,s,d,w)
```

```python
# Function to extract storage information
def extract_storage(title):
    if '- (' in title:
        return title.split('- ')[1].split('/')[1].strip()
    else:
        return None

# Fill null values in Storage column
df.loc[df['Storage'].isna(), 'Storage'] = df[df['Storage'].isna()]['Title'].apply(extract_storage)
```
