```python
def extract_features(soup):
    '''
    This function extracts the all features of the laptop from the soup object.
    In last it will return the dataframe of the features in one page for each laptop.
    These are the features that we are extracting:
    1. Title            2. Processor
    3. Ram              4. OS
    5. Storage          6. Display
    7. Warranty         8. Discount Price
    9. Original Price   10. Delivery
    11. Note        12. Offer       13. Rating and Rating

    '''
    title = []
    processor = []
    ram = []
    os = []
    storage = []
    display = []
    warranty = []


    discount_price = []
    original_price = []
    delivery = []
    note = []
    offer = []
    for i in soup.find_all('div',class_="yKfJKb row"):
        #title
        try:
            title.append(i.find_all('div',class_='KzDlHZ')[0].text)
        except:
            title.append(np.nan)
        
        #processor
        try:
            processor.append(i.find_all('li',class_='J+igdf')[0].text)
        except:
            processor.append(np.nan)
        
        #ram
        try:
            ram.append(i.find_all('li',class_='J+igdf')[1].text)
        except:
            ram.append(np.nan)
        
        #os
        try:
            os.append(i.find_all('li',class_='J+igdf')[2].text)
        except:
            os.append(np.nan)
        
        #storage
        try:
            storage.append(i.find_all('li',class_='J+igdf')[3].text)
        except:
            storage.append(np.nan)
        
        #display
        try:
            display.append(i.find_all('li',class_='J+igdf')[4].text)
        except:
            display.append(np.nan)
        
        #warrenty
        try:
            warranty.append(i.find_all('li',class_='J+igdf')[5].text)
        except:
            warranty.append(np.nan)

        #discount_price
        try:
            discount_price.append(i.find('div',class_ = "Nx9bqj _4b5DiR").text)
        except:
            discount_price.append(np.nan)
        
        #original_price
        try:
            original_price.append(i.find('div',class_ = "yRaY8j ZYYwLA").text)
        except:
            original_price.append(np.nan)
        
        #delivery
        try:
            delivery.append(i.find('div',class_ ="yiggsN").text)
        except:
            delivery.append(np.nan)
        
        #note
        try:
            note.append(i.find_all('div',class_ = "n5vj9c")[0].text)
        except:
            note.append(np.nan)
        
        #offer
        try:
            offer.append(i.find_all('div',class_ = "n5vj9c")[1].text)
        except:
            offer.append(np.nan)

    reviews = []
    for i in soup.find_all('div',class_="_5OesEi"):
        if i:
            reviews.append(i.text.replace('\xa0',' '))
        else:
            reviews.append(np.nan)
    while len(reviews)< len(title) :
        reviews.append(np.nan)
    # debugging
    print(len(title),len(processor),len(ram),len(os),len(storage),len(display),len(warranty),len(discount_price),len(original_price),len(delivery),len(note),len(offer),len(reviews))       
    
    # creating the dataframe
    df = pd.DataFrame({
        'Title':title,
        'Processor':processor,
        'Ram':ram,
        'OS':os,
        'Storage':storage,
        'Display':display,
        'Warranty':warranty,
        'Discount Price':discount_price,
        'Original Price':original_price,
        'Delivery':delivery,
        'Note':note,
        'Offer':offer,
        'Reviews':reviews
    })

    return df
```

```python
data = pd.DataFrame()
for i in range(1,42):
    print('Page:',i)
    url = f'https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    # debugging
    print(data.shape)
    data = pd.concat([data,extract_features(soup)],axis = 0,ignore_index = True)
    
    print('Data Extracted from page:',i)
    
    
```

```python
data.to_csv('laptop_data.csv',index = False)
```
